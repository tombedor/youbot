```python
from openai import OpenAI
from youbot.clients.memgpt_client import MemGPTClient
from youbot.onboard_user import onboard_user
from youbot.store import Store


PERSONA = """
My name is Alex Turner, and I'm 32 years old. I have a few hobbies that I’m really passionate about. I love photography, especially capturing urban landscapes and candid moments. On weekends, I often explore the city with my camera in hand, always on the lookout for interesting scenes. I also enjoy playing the guitar, a hobby I picked up in college. Evenings are often spent strumming and composing tunes. Gardening is another favorite pastime of mine; tending to my small but thriving collection of indoor plants brings me a lot of joy and relaxation.

I have a loyal and energetic Border Collie named Max. He loves playing fetch in the park and accompanies me on my weekend photography adventures. His playful and affectionate nature adds a lot of happiness to my life.

I’m in a loving relationship with Jordan, a graphic designer who shares my love for the arts. Jordan's creativity complements my photography, and we often collaborate on creative projects, blending my photos with Jordan's design skills.

I work as a marketing manager at a mid-sized tech company. The workplace is vibrant and dynamic, with a mix of enjoyable and challenging coworkers. I get along well with Sarah and Mike, who are always supportive and collaborative, making team projects a pleasure. However, dealing with Emily, who often misses deadlines, and Tom, who tends to micromanage, can be frustrating.

I’m excited about an upcoming trip to Kyoto, Japan. I plan to spend two weeks exploring the city, visiting historic temples, indulging in local cuisine, and of course, taking lots of photographs. This trip is a much-anticipated break from the daily grind and an opportunity to experience a different culture and gather inspiration for future creative projects.
"""

client = OpenAI()

class Chat:
    def __init__(self):
        self.messages = [{"role": "system", "content": "You are the following persona. Converse about topics relevant to your persona in character as though you were chatting with an assistant. Here is your persona: {PERSONA}\nFrom now on, act like your persona. You might say things like, 'Today I went to the beach with my dog Max'. Or 'I had a rough day at work today'."}]
        phone = "+17512383847"
        try:
            onboard_user(phone=phone, human_name="Alex", human_description="A new user")
        except ValueError:
            pass
        self.youbot_user = Store().get_youbot_user_by_phone(phone)
    
    def run(self, limit = 10):
        # we need to flip the user/assistant designation
        msgs = 0
        
        self.messages.append({"role": "assistant", "content": "Hi Alex, what did you do today?"})
        
        while msgs < limit:
            user_response = client.chat.completions.create(messages=self.messages, model='gpt-4').choices[0].message.content
            
            self.messages.append({"role": "user", "content": user_response})
            print(user_response)
            
            msgs += 1
            
            assistant_response = MemGPTClient().user_message(self.youbot_user, user_response)
            self.messages.append({"role": "system", "content": assistant_response})
            
            print(assistant_response)
        
        
        
Chat().run()
        
        
        
        
        
        
```
